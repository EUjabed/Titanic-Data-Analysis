```python
import pandas as pd
```

```python
data = pd.read_csv("train.csv")
test =pd.read_csv("test.csv")
```

```python
data.head()
```

```python
test.head()
```

```python
def clean(d):
    d = d.drop(["Ticket","Cabin","Name"], axis=1)
    cols = ["SibSp", "Parch", "Fare", "Age"]
    for col in cols:
        d[col].fillna(d[col].median(), inplace=True)
    d.Embarked.fillna("U", inplace=True)
    return d
```

```python
main = clean(data)
```

```python
main.head()
```

```python
new_test = clean(test)
```

```python
new_test.head()
```

```python
main.to_csv("main.csv", index=False)
```
